```python
import pandas as pd

# Charger le fichier CSV
df = pd.read_csv("./data/patients.csv")

# Afficher les premières lignes
df.head()
```

```python
import matplotlib.pyplot as plt

df['Age'].hist()
plt.title("Distribution des âges des patients")
plt.xlabel("Âge")
plt.ylabel("Nombre de patients")
plt.show()
```

```python

# Run basic descriptive statistics using:

df.describe()

df['Sex'].value_counts()

df['Diagnosis'].value_counts()
```

```python
# Visualize the data: Create a Simple Plot using matplotlib

import matplotlib.pyplot as plt

plt.plot(df["Age"], df["Cholesterol"], marker="o")

plt.xlabel("Age")

plt.ylabel("Cholesterol")

plt.title("Cholesterol vs Age")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Âge selon le diagnostic
sns.boxplot(x="Diagnosis", y="Age", data=df)
plt.title("Âge selon le diagnostic")
plt.show()

# Cholestérol selon le sexe
sns.boxplot(x="Sex", y="Cholesterol", data=df)
plt.title("Cholestérol selon le sexe")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Matrice de corrélation")
plt.show()
```
